```python
# Importation of libraries
import numpy as np
import csv
import pandas as pd
```

```python
# Importation of databases
database = pd.read_csv("database.csv", delimiter = ';')
result = pd.read_csv("result.csv", delimiter = ',')
```

```python
result.head()
```

```python
result.drop(columns=['Unnamed: 0'], inplace = True)
```

```python
result.head()
```

```python
# In the database result, we had a column with the winner of the matchup Poke1 vs Poke2
result.loc[result['Nb_victory_pok1'] == result['Nb_victory_pok2'], 'Winner'] = 'draw'
result.loc[result['Nb_victory_pok1'] > result['Nb_victory_pok2'], 'Winner'] = result['Pokemon1']
result.loc[result['Nb_victory_pok1'] < result['Nb_victory_pok2'], 'Winner'] = result['Pokemon2']
```

```python
result.head()
```

```python
# We create a new dataset which contains the number of victory of each Pokemon
count = {
    'name_poke' : [],
    'nb_vic' : []
    }

df_count = pd.DataFrame(count)

for i in range(len(database)):
    a = database["name_poke"][i]
    b = result[result["Winner"]==database["name_poke"][i]].count()["Pokemon1"]
    df_count.loc[i] = [a,b]

df_count
```

```python
# Now, we can merge our dataframes to add the new column of the database : nb_vic 
database2 = pd.merge(database, df_count, how='left', on="name_poke")
```

```python
# The last step is to add the column win_rate to the enrich_database
database2["win_rate"] = (database2["nb_vic"]/len(database2)).round(2)
database2.head()
```

```python
database2.to_csv('database2.csv')
```
